```python
#Set
setmobil = {"Toyota", "Honda", "BMW", "Mercedes"}
print(setmobil)
print(type(setmobil))
```

```python
#akses elemen dalam set
for x in setmobil:
    print(x)    
```

```python
#Menambah elemen set dengan add
setmobil.add("Audi")
setmobil.add("Lexus")
setmobil.add("Volkswagen")
print(setmobil) 
```

```python
#update
setmobil.update(["Nissan", "Mazda"])
print(setmobil)
```

```python
#mengukur jumlah elemen dalam set
print(len(setmobil))
```

```python
#menghapus elemen set remove
setmobil.remove("BMW")
print(setmobil)
```

```python
#menghapus elemen set dengan discard
setmobil.discard("Mercedes")
print(setmobil)
```

```python
#menghapus elemen set denagn pop
setmobil.pop()
print(setmobil)
```

```python
print(setmobil)
```

```python
#menghapus element set dengan clear
setmobil.clear()
print(setmobil)
```

```python
#menghapus elemen set dengan del
del setmobil
print(setmobil)
#Hasil pasti error karena setmobil sudah dihapus dengan del(SECARA KESELURUHAN)
```

```python
#menggabungkan set
mobilcina = {"Geely", "Wuling", "Chery"}
mobiljepang = {"Toyota", "Honda", "Nissan"}

#menggunakan command union
gabungan = mobilcina.union(mobiljepang)
print(gabungan)

#menggunakan update
mobilcina.update(mobiljepang)
print(mobilcina) 
```

```python
#DICTIONARY
daftarapp = {"app" : "Instagram"
             "app2" : "Facebook
             "app3" : "Twitter"}
print(daftarapp)
print(type(daftarapp))
```

```python
#akses element dictionary
x = daftarapp["app"]
print(x)
```

```python
#Menggunakan set dictionary
v = daftarapp.get("app2")
print(v)
```

```python
#contoh loop
for x in daftarapp:
    print(daftarapp[x])
```

```python
#contoh loop 2
for x in daftarapp.values():
    print(x)  
```

```python
#contoh loop 3
for x, y in daftarapp.items():
    print(x, y)
```

```python
#cek apakah key tertentu ada dalam dictionary
if "Tahun" in daftarapp:
    print("Key 'Tahun' ditemukan dalam dictionary")   
else:
    print("Key 'Tahun' tidak ditemukan dalam dictionary")
```

```python
#menentukan jumlah elemen dalam dictionary
print(len(daftarapp))
#Yg dihitung itu jumlah pasangan key-value, bukan jumlah key berapa ditambah value berapa. Jadi, dalam contoh ini, ada 3 pasangan key-value, sehingga hasilnya adalah 3.
```

```python
#menambahkan item kedalam dictionary
daftarapp["app4"] = "TikTok"
print(daftarapp)
```

```python
#menghapus item dengan command pop()
daftarapp.pop("app2")
print(daftarapp)
```

```python
#menggunakan popitem()
daftarapp.popitem()
print(daftarapp)
```

```python
#menghapus elemen dictionary dengan clear
daftarapp.clear()
print(daftarapp)
```

```python
#menghapus elemen dictionary dengan del
del daftarapp
print(daftarapp)
#Hasil pasti error karena daftarapp sudah dihapus dengan del(SECARA KESELURUHAN)
```

```python
#menyalin dictionary dengan copy
daftarapp2 = daftarapp.copy()
print(daftarapp2)
```

```python
#menyalin dictionary dengan dict
daftarapp3 = dict(daftarapp)
print(daftarapp3)
```

```python
#nested dictionary
daftarapp4 = {
    "sosmed" : {
        "instagram" : "user1",
        "twitter" : "user2"
    }
    "game" : {
        "mobile legends" : "player1",
        "pubg" : "player2"
    }
}
print(daftarapp4)
```
