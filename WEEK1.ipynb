RANDOM

```python
import random

# 1. Angka Bulat (INTEGER)
hasil_dadu = random.randint(1, 6)
print("Lempar dadu: ", hasil_dadu)

# 2. Angka Genap Acak
genap = random.randrange(0, 11, 2)
print("Angka genap acak: ", genap)

# 3. Angka desimal(Float)
probabilitas = random.random()
print("Nilai probabilitas: ", probabilitas)

# 4. Angka desimal dalam rentang
suhu = random.uniform(1.5, 5.5)
print("suhu acak: ", suhu)

# 5. Pilih 1 Elemen acak
nama = random.choice(['arip', 'leon', 'elpointer', 'elvolume'])
print("Nama terpilih: ", nama)
nama = random.choice(['arip', 'leon', 'elpointer', 'elvolume'])
print("Nama terpilih: ", nama)

# 6. Acak posisi elemen
my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)
print("List setelah diacak: ", my_list)

# 7. Ambil K elemen ack

# 8. Pengambilan dengan bobot
hadiah = ['mobil', 'motor', 'sepeda', 'uang']
peluang = [5, 20, 50, 25]
pemenang = random.choices(hadiah, weights=peluang, k=1)
print("Pemenang hadiah: ", pemenang[0])
```

OPERATOR ARITMATIKA

```python
#VARIABEL
x = 5
y = 10

#1. Penjumlahan variabel
z = x + y
print("Hasil penjumlahan x dan y adalah: ", z)

#2. Pengurangan variabel
z = x - y
print("Hasil pengurangan x dan y adalah: ", z)

#3. Perkalian variabel
z = x * y
print("Hasil perkalian x dan y adalah: ", z)

#4. Pembagian variabel
z = x / y
print("Hasil pembagian x dan y adalah: ", z)

#5. modulus variabel
z = x % y
print("Hasil modulus x dan y adalah: ", z)

# 6. Eksponen variabel
z = x ** 2
print("Hasil eksponen x adalah: ", z)

#7. Floor division variabel
z = x // y
print("Hasil floor division x dan y adalah: ", z)
```

OPERATOR PERBANDINGAN

```python
a = 10
b = 20

#1. Sama dengan (==)
print("Apakah a sama dengan b? ", a == b)

#2. Tidak sama dengan (!=)
print("Apakah a tidak sama dengan b? ", a != b)

#3. Lebih besar dari (>)
print("Apakah a lebih besar dari b? ", a > b)

#4. Lebih kecil dari (<)
print("Apakah a lebih kecil dari b? ", a < b)

#5. Lebih besar atau sama dengan (>=)
print("Apakah a lebih besar atau sama dengan b? ", a >= b)

#6. Lebih kecil atau sama dengan (<=)
print("Apakah a lebih kecil atau sama dengan b? ", a <= b)

```

OPERATOR LOGICAL

```python
#OPERATOR LOGICAL

#1 AND
a, b = True, True
print("Hasil a AND b: ", a and b)

#2. OR
print("Hasil a OR b: ", a or b)

#3. NOT
print("Hasil NOT a: ", not a)

#4. XOR
print("Hasil a XOR b: ", a ^ b)

#5. NAND
print("Hasil a NAND b: ", not (a and b))

#6. NOR
print("Hasil a NOR b: ", not (a or b))

#7. Left Shift
print("Hasil a Left Shift 1: ", a << 1)

#8. Right Shift
print("Hasil a Right Shift 1: ", a >> 1)
```

OPERATOR IDENTYTY

```python
#OPERATOR IDENTITY
a, b = [1, 2, 3], [1, 2, 3]

#1. IS
print("Apakah a dan b adalah objek yang sama? ", a is b) #Ngecek apakah a dan b adalah objek yang sama, bukan hanya nilai yang sama. Karena a dan b adalah list yang berbeda, maka hasilnya adalah False.

#2. IS NOT
print("Apakah a dan b adalah objek yang berbeda? ", a is not b)
```
