```python
#TIPE DATA BOOLEAN
print(True)
print(False)
print(type(True))
print(type(False))
```

```python
#TIPE DATA STRING
print("Belajar python mudah dan menyenangkan bukan")
print(type("Belajar python mudah dan menyenangkan bukan"))
```

```python
#TIPE DATA INTEGER
print(10)
print(type(10))
```

```python
#TIPE DATA FLOAT
print(3.14)
print(type(3.14))
```

```python
#TIPE DATA COMPLEX
print(2 + 3j)
print(type(2 + 3j))
```

```python
#TIPE DATA HEKSADESIMAL
print(0x27)
print(type(0x27))
```

```python
#TIPE DATA LIST
print([1, 2, 3, 4, 5])
print([1, "dua", 3.0, True])
print(type([1, 2, 3, 4, 5]))
print(type([1, "dua", 3.0, True]))
```

```python
#TIPE DATA TUPLE
print((1, 2, 3, 4, 5))
print((1, "dua", 3.0, True))
print(type((1, 2, 3, 4, 5)))
print(type((1, "dua", 3.0, True)))
```

```python
#TIPE DATA SET
print({1, 2, 3, 4, 5})
print({1, "dua", 3.0, True})
print(type({1, 2, 3, 4, 5}))
print(type({1, "dua", 3.0, True}))
```
