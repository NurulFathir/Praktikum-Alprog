MODUL 2 PYTHON COLLECTION(ARRAY)

```python
#Membuat List
thislist = ["apple", "banana", "cherry"]
print(thislist)
```

```python
#Mengakses Item list(Positif index)
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```

```python
#Mengakses Item list(Negatif index)
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

```python
#mencari posisi item dengan index()
thislist = ["apple", "banana", "cherry"]
postion = thislist.index("banana")
print(postion)
```

```python
#Mencari posisi item dengan index() pada item yang duplikat
thislist = ["apple", "banana", "cherry", "banana"]
postion = thislist.index("banana")
print(postion)
```

```python
#mengambil range item dengan slicing
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
print(thislist[:4])
print(thislist[2:])
```

```python
#Mengambil range item list(negatif index)
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

```python
#Mengubah Item List
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

```python
#Perulangan list
thislist = ["apple", "banana", "cherry"]
for x in thislist:
  print(x)
```

```python
#Mengecek apakah item tertentu ada dalam list
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("YAP, 'apple' ADA DALAM LIST")
```

```python
#Menghitung jumlah item dalam list
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

```python
#Menghitug jumlah kemunculan satu nilai spesifik
thislist = ["apple", "banana", "cherry", "apple", "apple"]
position = thislist.count("apple")
print(position)
```

MENAMBAKAN ITEM KE LIST

```python
#append() Menambahkan item ke akhir list
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
#insert() Menambahkan item pada posisi tertentu
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)
```

MENGHAPUS ITEM DI LIST

```python
#Remove() Menghapus item tertentu
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)
```

```python
#pop() untuk menghapus item pada posisi tertentu
thislist = ["apple", "banana", "cherry"]
diambil = thislist.pop(1)
print("Item yg diamnbil: ", diambil)
print("List setelah penghapusan: ", thislist)
```

```python
#del Remove item pada posisi tertentu
thislist = ["apple", "banana", "cherry"]
del thislist[0]
print("List setelah penghapusan: ", thislist)
```

```python
#Clear() Mengosongkan seluruh list dengan
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print("List setelah pengosongan: ", thislist)
```

```python
#menghapuys seluruh list
thislist = ["apple", "banana", "cherry"]
del thislist
print(thislist)
```

```python
#mengcopy list menggunakan = 
list1 = ["apple", "banana", "cherry"]
mylist = list1
print(mylist)

list1[0] = "orange"
list1[1] = "melon"
print("list1: ", list1)
print("mylist: ", mylist)
```

```python
#mengcopy list menggunakan copy()
list1 = ["apple", "banana", "cherry"]
mylist = list1.copy()
print(mylist)

list1[0] = "orange"
list1[1] = "melon"
print("list1: ", list1)
print("mylist: ", mylist)
```

```python
#contoh lain mengcopy list
list1 = ["apple", "banana", "cherry"]
mylist = list(list1)
print(mylist)

list1[0] = "orange"
print("list1: ", list1)
print("mylist: ", mylist)
```

MENGGABUNGKAN DUA LIST

```python
#menggunaan operator + untuk menggabungkan list
list1 = ["apel", "lontar" , "cherry"]
list2 = ["Mobil", "Motor", "Truk"]
list3 = list1 + list2
print(list3)
```

```python
#menggunakan append() untuk menggabungkan list
list1 = ["apel", "lontar" , "cherry"]
list2 = ["Mobil", "Motor", "Truk"]
for x in list2:
  list1.append(x)
print(list1)
```

```python
#menggunakan extend() untuk menggabungkan list
list1 = ["apel", "lontar" , "cherry"]
list2 = ["Mobil", "Motor", "Truk"]
list1.extend(list2)
print(list1)
```

```python
#Membuat list dengan constructor
thislist = list(("apple", "banana", "cherry"))
print(thislist)
```

MENGURUTKAN LIST

```python
#reverse urutan list menggunakan reverse()
thislist = ["apple", "banana", "cherry"]
thislist.reverse()
print(thislist)
```

```python
#Mengurutkan list menggunakan sort()
thislist = ["apple", "banana", "cherry"]
thislist.sort()
print("Ascending: ", thislist)
thislist.sort(reverse = True)
print("Descending: ", thislist)

strlist = ["banana", "cherry", "apple"]
strlist.sort()
print("Sorted list: ", strlist)
strlist.sort(reverse = True)
print("reverse sorted list: ", strlist)
```

```python
#mengurutkan list dengan fungsi sorted()
numlist = [1, 5, 3, 2, 4]
sortedlist = sorted(numlist)
print("Original list: ", numlist)
print("Sorted list: ", sortedlist)
```

```python
#mengurutkan dengan key(custom sorting)
thislist = ["apple", "key", "cherry", "kiwi", "mango"]
thislist.sort(key = len)
print("Sorted by length: ", thislist)
```
