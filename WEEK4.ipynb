NESTED IF

```python
x = 41

if x > 10:
    print("x lebih besar dari 10")
    if x > 10 and x < 50:
        print("x lebih besar dari 10 dan kurang dari 50")
    else:
        print("x lebih besar dari 10 dan lebih besar atau sama dengan 50")
```

WHILE

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

WHILE BREAK

```python
i = 1
while i <= 5:
    print(i)
    if i == 3:
        break
    i += 1
```

WHILE CONTIONUE

```python
i = 1
while i <= 5:
    print(i)
    if i == 3:
        continue
    i += 1
```

WHILE ELSE

```python
i = 1
while i <= 5:
    print(i)
    i = i + 1
else:
    print("Looping selesai")
```

FOR 

```python
buah = ["apel", "jeruk", "pisang"]

for b in buah:
    print(b)
```

FOR STRING

```python
for x in "elektro"
    print(x)
```

FOR BREAK

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
```

FOR CONTINUE

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        continue
    print(x)
```
