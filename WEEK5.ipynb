MEMBUAT ARRAY SLICING

```python
#NumPy array slicing 1 dimensi

import numpy as np
e = np.array([1, 3, 6, 8, 2, 3])
print(e[1:4])

print(e[:5])

print(e[-3: 1])

print(e[::2])
```

TIPE DATA PADA NumPy

```python
import numpy as np
a = np.array([1, 2, 3])
print(a.dtype)

b = np.array(["MBG", "19 juta lapker", "wifi gratis"])
print(b.dtype)
```

```python
#Membuat array dngn tipe data yang ditentukan
import numpy as np
a = np.array([1, 3, 4], dtype='S')
print(a)
print(a.dtype)

print(' ')
b = np.array([1, 3, 4], dtype='i')
print(b)
print(b.dtype)

#dibagian ini cantumin analisisnya nanti di laprak, kenapa bsa gini
```

```python
c = np.array(['a', 3, 4, 5, 1], dtype = 'i')
print(c)
print(c.dtype)

#ini bakal eror karna a itu ga bsa diubah jadi integer
#kecuali string yg isinya angka baru bsa diubah jdi integer
```

```python
#mengoncersi jenis data pada tipe yg ada
import numpy as np
a = np.array([1, 2, 3, 4, 5])
a_new = a.astype('i')
print(a)
print(a.dtype)
print(a_new)
print(a_new.dtype)

print(' ')
a_new2 = a.astype(int)
print(a_new2)
print(a_new2.dtype)

print(' ')
b = np.array([2, 4, 5,1, 0, -1, -2]) #hanya melihat 1 sebagai true dan 0 sebagia false
bool_b = b.astype(bool)
print(bool_b)
print(bool_b.astype)
```

PERBEDAAN COPY DAN VIEW

```python
#konversi jenis data pada tipe data yg ada
import numpy as np
arrx = np.array([1, 2, 3, 4, 5])
x = arrx.copy()#membuat salinan dari arrx
arrx[0] = 12 #membuat nilai dalam array
print(x)
print(arrx)

print(' ')
arry = np.array([1, 2, 3, 4, 5])
y = arry.view()
arry[0] = 18
print(y)
print(arry)

print(' ')
print(x.base)
print(y.base)
```
