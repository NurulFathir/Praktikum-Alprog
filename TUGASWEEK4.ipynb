TUGAS: KONVERSI NILAI

```python
skor = int(input("Masukan Skor :"))

if skor >= 80 and skor <= 100:
    print("A")
elif skor >= 75:
    print("B")
elif skor >= 60:
    print("C")
elif skor >= 35:
    print("D")
else:
    print("E")
```

TUGAS FAKTORIAL 10!

```python
faktorial = 1
i = 1

while i <= 10:
    faktorial *= i
    i += 1

print("Faktorial dari 10 adalah:", faktorial)
```
